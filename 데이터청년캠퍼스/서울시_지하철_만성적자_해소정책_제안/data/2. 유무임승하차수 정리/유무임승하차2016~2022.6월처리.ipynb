```python
import pandas as pd

df = pd.read_csv('역별유무임(1~9호선).csv', encoding='cp949')
df = df[df['사용월']//100 != 2015]
# 15년 데이터 수치 이상으로 제외
```

```python
df['호선명'] = df['호선명'].str.replace('2~3단계','')
df['호선명'] = df['호선명'].str.replace('2단계','')

for i in range(1,10):
    df[df['호선명']==str(i)+'호선']
    
my_dict = {'유임승차인원':'mean', '무임승차인원':'mean','유임하차인원':'mean','무임하차인원':'mean'}
```

```python
a = df[df['호선명']==str(1)+'호선'].groupby('지하철역').agg(my_dict)
a.insert(0, '호선', '1호선')

for i in range(2,10):
    b = df[df['호선명']==str(i)+'호선'].groupby('지하철역').agg(my_dict)
    b['호선']=str(i)+'호선'
    a = pd.concat([a,b])

a
```
